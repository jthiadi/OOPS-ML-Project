```python
# === YOLO Webcam Inference with "unknown" label support ===
from ultralytics import YOLO
import cv2, torch

# ---------------- CONFIGURATION ----------------
MODEL_PATH = "best_3.pt"   # your trained model
CONF_FILTER = 0.3           # YOLO internal filter (ignore detections below this)
UNKNOWN_THRESHOLD = 0.6    # if conf < this, label = "unknown"
DEVICE = 0 if torch.cuda.is_available() else "cpu"

# ---------------- LOAD MODEL ----------------
model = YOLO(MODEL_PATH)
print(f"âœ… Model loaded on device: {DEVICE}")

# ---------------- OPEN WEBCAM ----------------
cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)
if not cap.isOpened():
    raise RuntimeError("âŒ Could not open webcam. Try a different index (1/2).")

print("ðŸŽ¥ Webcam running â€” press 'q' to quit.")

# ---------------- MAIN LOOP ----------------
while True:
    ret, frame = cap.read()
    if not ret:
        print("âš ï¸ Frame not captured, exiting.")
        break

    # Run YOLO prediction on current frame
    results = model.predict(
        source=frame,
        conf=CONF_FILTER,
        iou=0.45,
        device=DEVICE,
        verbose=False
    )

    annotated = frame.copy()

    # Draw bounding boxes with "unknown" label support
    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls = int(box.cls[0])

            if conf < UNKNOWN_THRESHOLD:
                label = "unknown"
            else:
                label = f"{model.names[cls]} ({conf:.2f})"

            cv2.rectangle(annotated, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(annotated, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Display the live feed
    cv2.imshow("YOLO Webcam (press 'q' to quit)", annotated)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print("ðŸ‘‹ Webcam closed.")
```
